```python
#Python Code Three Day Chalenge 
#Team 20
#Member's:-
#Hiren Rathod | hiren.rathod@epita.fr
#Prateek AREKALLU PRAFULLA CHANDRA | prateek.arekallu-prafulla-chandra@epita.fr
#Vinay DARAPUREDDI | vinay.darapureddi@epita.fr


#import part
import time
import sys
import os
from goto import with_goto
#global values
load = 0                                 # set initial loading value
walk = [(1,0),(-1,0),(0,1),(0,-1)]       # set movements
backtracker = []                         # set backtracker (remember that this is a stack)
gmainmaze = []                           # set initial blank maze
doorTracker = []
FinalTracker = []

def initialize():   #loading 
    sx = 0 # start x value
    sy = 0 # start y value
    kx = 0 
    ky = 0
    dx = 0
    dy = 0
    ex = 0 # end x value
    ey = 0 # end y value
    while load == 0:    #just some cool animation
        sys.stdout.write("\rloading |")
        time.sleep(0.1)
        sys.stdout.write("\rloading /")
        time.sleep(0.1)
        sys.stdout.write("\rloading -")
        time.sleep(0.1)
        sys.stdout.write("\rloading \\")
        time.sleep(0.1)
        sys.stdout.write("\r")
        mainmaze = loadMaze("Maze2.txt")
        print("Maze Loaded")
        print(" ")
        time.sleep(1)
        print(" ")
        print("Locating start point...")
        sx, sy = spotter("s",mainmaze)
        time.sleep(1)
        print(" ")
        print("Locating key point...")
        kx, ky = spotter("a",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating door point...")
        dx, dy = spotter("b",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating end point...")
        ex, ey = spotter("e",mainmaze)
        print(" ")
        time.sleep(1)
        return sx, sy, kx, ky, dx, dy, ey, ex, mainmaze;
        break
    print(" ")
    sys.stdout.write("\rInitialized!")
    print(" ")
    time.sleep(3)
    os.system("cls")

def loadMaze(filename):
    #load the data from the file into a 2D list
    with open(filename) as i:
        maze = []
        for line in i:
            line = line.strip()
            spawnmaze = line.split(", ")
            maze.append(spawnmaze)
    return maze

def spotter(num,maze):
    #Locate the 'element' in the maze (this can either be "s" or "e")
    num = str(num)
    rowcounter = -1
    linecounter = 0
    for rows in maze:
        rowcounter = rowcounter + 1
        if num in rows:
            for element in rows:
                if element == num:
                    print("Tango Spotted, Grid:", rowcounter, linecounter)
                    return rowcounter, linecounter;
                    break
                linecounter = linecounter + 1

def valid(maze,x,y): # check if valid
    height = len(maze) - 1
    width = len(maze[0]) - 1
    if 0 <= x <= height and 0 <= y <= width:
        return 1
    else: 
        return 0

   
def findKey(sx,sy,kx,ky): 
    global gmainmaze   
    for i in walk:          #try every direction
        x = sx + i[0]       #make the move
        y = sy + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == kx and y == ky: #check if reached destination
                print("Destination Arrived : key Found")
                print("The Logistic Path Is:\n",backtracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in backtracker: #add to the stack
                    backtracker.append((x,y))
               
                else:
                    continue
        else:
            continue
        
        if findKey(x,y,kx,ky) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    backtracker.pop(-1) #moveback while 
    return 0


def keyToDoor(kx,ky,dx,dy):
    global gmainmaze   
    for i in walk:          #try every direction
        x = kx + i[0]       #make the move
        y = ky + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == dx and y == dy: #check if reached destination
                print("Destination Arrived: Door Found")
                print("The Logistic Path Is:\n",doorTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in doorTracker:
                    doorTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if keyToDoor(x,y,dx,dy) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    doorTracker.pop(-1) #moveback while 
    return 0

def DoorToDestination(dx,dy,ex,ey):
    global gmainmaze   
    for i in walk:          #try every direction
        x = dx + i[0]       #make the move
        y = dy + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == ex and y == ey: #check if reached destination
                print("Destination Arrived: Door Unlocked and Reached Destination")
                print("The Logistic Path Is:\n",FinalTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in FinalTracker:
                    FinalTracker.append((x,y)) 
                else:
                    continue
        else:
            continue
        
        if DoorToDestination(x,y,ex,ey) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    FinalTracker.pop(-1) #moveback while 
    return 0
    
def printMaze(maze):
    for x in range(len(maze)):
        for y in range(len(maze[x])):
            print(maze[x][y], end=' ')
        print("")

def visualize(maze):    # a cool function to visualize the maze
    for x in range(len(maze)):
        for y in range(len(maze[x])):
            if maze[x][y] == "s":
                maze[x][y] = "╳"    # path
            elif maze[x][y] == "0":
                maze[x][y] = "░"    # unbeen path
            elif maze[x][y] == "1":
                maze[x][y] = "█"    # wall
            elif maze[x][y] == "e":
                maze[x][y] = "╳"    # basically path as well
            print(maze[x][y], end=' ')
        print("")

     
def main():
    #initialize the maze
    sx, sy, kx, ky, dx, dy, ex, ey, mainmaze = initialize()
    global gmainmaze
    gmainmaze = mainmaze
    load = 1
    #solve the maze
    findKey(sx,sy,kx,ky)
    keyToDoor(kx,ky,dx,dy)
    DoorToDestination(dx,dy,ex,ey)
    for i in backtracker:
        gmainmaze[i[0]][i[1]] = "s"

    for i in doorTracker:
        gmainmaze[i[0]][i[1]] = "s"
 
    for i in FinalTracker:
        gmainmaze[i[0]][i[1]] = "s"
    
    #print the maze
    width = len(gmainmaze[0]) - 1
    print("Check Your Map...")
    print("Map:")
    print("--"*(width+1))
    printMaze(gmainmaze)
    print("--"*(width+1))
    #visualize the maze uncomment to establish the function (btw check if your terminal can print these unicodes)
    visualize(gmainmaze)
    time.sleep(1)

main()
```
