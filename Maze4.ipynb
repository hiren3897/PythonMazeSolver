```python
#Python Code Three Day Chalenge 
#Team 20
#Member's:-
#Hiren Rathod | hiren.rathod@epita.fr
#Prateek AREKALLU PRAFULLA CHANDRA | prateek.arekallu-prafulla-chandra@epita.fr
#Vinay DARAPUREDDI | vinay.darapureddi@epita.fr

#import part
import time
import sys
import os
from goto import with_goto
#global values
load = 0                                 # set initial loading value
walk = [(1,0),(-1,0),(0,1),(0,-1)]       # set movements
backtracker = []                         # set backtracker (remember that this is a stack)
gmainmaze = []                           # set initial blank maze
keyTracker = []                          #it will cantain all the logistics of key path
YdoorTracker = []
GdoorTracker = []
RdoorTracker = []
BdoorTracker = []
FinalTracker = []

def initialize():   #loading 
    sx = 0   # start x value
    sy = 0   # start y value
    ykx = 0  # Yellow Key X value
    yky = 0  # Yellow Key Y value
    ydx = 0  # Yellow Door X value
    ydy = 0  # Yellow Door Y value
    gkx = 0  # Green Key X value
    gky = 0  # Green Key Y value
    gdx = 0  # Green Door X value
    gdy = 0  # Green Door Y value
    rkx = 0  # Red Key X value
    rky = 0  # Red Key Y value
    rdx = 0  # Red Door X value
    rdy = 0  # Red Door Y value
    bkx = 0  # Blue Key X value
    bky = 0  # Blue Key Y value
    bdx = 0  # Blue Door X value
    bdy = 0  # Blue Door Y value
    ex = 0   # end x value
    ey = 0   # end y value
    devilx = 0;
    devily = 0;
    while load == 0:    #just some cool animation
        sys.stdout.write("\rloading |")
        time.sleep(0.1)
        sys.stdout.write("\rloading /")
        time.sleep(0.1)
        sys.stdout.write("\rloading -")
        time.sleep(0.1)
        sys.stdout.write("\rloading \\")
        time.sleep(0.1)
        sys.stdout.write("\r")
        mainmaze = loadMaze("Maze4.txt")
        print("Maze Loaded")
        print(" ")
        time.sleep(1)
        print(" ")
        print("Locating start point...")
        sx, sy = spotter("s",mainmaze)
        time.sleep(1)
        print(" ")
        print("Locating Yellow key point...")
        ykx, yky = spotter("a",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Yellow door point...")
        ydx, ydy = spotter("b",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Green key point...")
        gkx, gky = spotter("c",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Green door point...")
        gdx, gdy = spotter("d",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Red key point...")
        rkx, rky = spotter("f",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Red door point...")
        rdx, rdy = spotter("g",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Blue key point...")
        bkx, bky = spotter("h",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Blue door point...")
        bdx, bdy = spotter("i",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating end point...")
        ex, ey = spotter("e",mainmaze)
        print(" ")
        time.sleep(1)
       
        print("Locating Devil 1...")
        devilx,devily = spotter("4",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Devil 2...")
        devilx,devily = spotter("2",mainmaze)
        print(" ")
        time.sleep(1)
        print("Locating Devil 3...")
        devilx,devily = spotter("3",mainmaze)
        print(" ")
        time.sleep(1)
        return sx, sy, ykx, yky, ydx, ydy, gkx, gky, gdx, gdy, rkx, rky, rdx, rdy, bkx, bky, bdx, bdy, ey, ex, mainmaze;
        break
    print(" ")
    sys.stdout.write("\rInitialized!")
    print(" ")
    time.sleep(3)
    os.system("cls")

# Loading The Maze from File by using delimeter ',' 
def loadMaze(filename):
    #load the data from the file into a 2D list
    with open(filename) as i:
        maze = []
        for line in i:
            line = line.strip()
            spawnmaze = line.split(", ")
            maze.append(spawnmaze)
    return maze

# Function Created for spotting Start, Key, Door, End point and Enemy
def spotter(num,maze):
    #Locate the 'element' in the maze
    num = str(num)
    rowcounter = -1
    linecounter = 0
    for rows in maze:
        rowcounter = rowcounter + 1
        if num in rows:
            for element in rows:
                if element == num:
                    print("Tango Spotted, Grid:", rowcounter, linecounter)
                    return rowcounter, linecounter;
                    break
                linecounter = linecounter + 1

def valid(maze,x,y): # check if valid
    height = len(maze) - 1
    width = len(maze[0]) - 1
    if 0 <= x <= height and 0 <= y <= width:
        return 1
    else: 
        return 0

   
def findKey(sx,sy,ykx,yky): 
    global gmainmaze   
    for i in walk:          #try every direction
        x = sx + i[0]       #make the move
        y = sy + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == ykx and y == yky: #check if reached destination
                print("Destination Arrived : key Found")
                print("The Logistic Path Is:\n",backtracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in backtracker: #add to the stack
                    backtracker.append((x,y))
               
                else:
                    continue
        else:
            continue
        
        if findKey(x,y,ykx,yky) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    backtracker.pop(-1) #moveback while 
    return 0


def YkeyToYDoor(ykx,yky,ydx,ydy):
    global gmainmaze   
    for i in walk:          #try every direction
        x = ykx + i[0]       #make the move
        y = yky + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == ydx and y == ydy: #check if reached destination
                print("Destination Arrived: Door Found")
                print("The Logistic Path Is:\n",YdoorTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in YdoorTracker:
                    YdoorTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if YkeyToYDoor(x,y,ydx,ydy) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    YdoorTracker.pop(-1) #moveback while 
    return 0

def GkeyToGDoor(a,b,c,d):
    global gmainmaze   
    for i in walk:          #try every direction
        x = a + i[0]       #make the move
        y = b + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == c and y == d: #check if reached destination
                print("Destination Arrived: Green Door Found")
                print("The Logistic Path Is:\n",GdoorTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in GdoorTracker:
                    GdoorTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if GkeyToGDoor(x,y,c,d) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    GdoorTracker.pop(-1) #moveback while 
    return 0

def RkeyToRDoor(a,b,c,d):
    global gmainmaze   
    for i in walk:          #try every direction
        x = a + i[0]       #make the move
        y = b + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == c and y == d: #check if reached destination
                print("Destination Arrived: Red Door Found")
                print("The Logistic Path Is:\n",RdoorTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in RdoorTracker:
                    RdoorTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if RkeyToRDoor(x,y,c,d) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    RdoorTracker.pop(-1) #moveback while 
    return 0

def BkeyToBDoor(a,b,c,d):
    global gmainmaze   
    for i in walk:          #try every direction
        x = a + i[0]       #make the move
        y = b + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == c and y == d: #check if reached destination
                print("Destination Arrived: Blue Door Found")
                print("The Logistic Path Is:\n",BdoorTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in BdoorTracker:
                    BdoorTracker.append((x,y))
                elif gmainmaze[x][y] == "h" and (x,y) not in BdoorTracker:
                    BdoorTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if BkeyToBDoor(x,y,c,d) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    BdoorTracker.pop(-1) #moveback while 
    return 0

def DoorToKey(a,b,c,d):
    global gmainmaze   
    for i in walk:          #try every direction
        x = a + i[0]       #make the move
        y = b + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == c and y == d: #check if reached destination
                print("Destination Arrived: Next Key Found")
                print("The Logistic Path Is:\n",keyTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in keyTracker:
                    keyTracker.append((x,y))
                    
                else:
                    continue
        else:
            continue
        
        if DoorToKey(x,y,c,d) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    keyTracker.pop(-1) #moveback while 
    return 0


def DoorToDestination(ydx,ydy,ex,ey):
    global gmainmaze   
    for i in walk:          #try every direction
        x = ydx + i[0]       #make the move
        y = ydy + i[1]
        if valid(gmainmaze,x,y): # check if still in the maze
            if x == ex and y == ey: #check if reached destination
                print("Destination Arrived: Door Unlocked and Reached Destination")
                print("The Logistic Path Is:\n",FinalTracker)
                print(" ")
                return 1
            else:
                if gmainmaze[x][y] == "0" and (x,y) not in FinalTracker:
                    FinalTracker.append((x,y)) 
                else:
                    continue
        else:
            continue
        
        if DoorToDestination(x,y,ex,ey) == 1: # Recursion (do the next step)
            return 1
        else:
            continue
    FinalTracker.pop(-1) #moveback while 
    return 0
    
def printMaze(maze):
    for x in range(len(maze)):
        for y in range(len(maze[x])):
            print(maze[x][y], end=' ')
        print("")

def visualize(maze):    # a cool function to visualize the maze
    for x in range(len(maze)):
        for y in range(len(maze[x])):
            if maze[x][y] == "s":
                maze[x][y] = "╳"    # path
            elif maze[x][y] == "0":
                maze[x][y] = "░"    # unbeen path
            elif maze[x][y] == "1":
                maze[x][y] = "█"    # wall
            elif maze[x][y] == "e":
                maze[x][y] = "╳"    # basically path as well
        
            print(maze[x][y], end=' ')
        print("")

     
def main():
    #initialize the maze
    sx, sy, ykx, yky, ydx, ydy, gkx, gky, gdx, gdy, rkx, rky, rdx, rdy, bkx, bky, bdx, bdy, ey, ex, mainmaze = initialize()
    global gmainmaze
    gmainmaze = mainmaze
    load = 1
    #solve the maze
    #gmainmaze[sx][sy] = "0"
    #gmainmaze[kx][ky] = "0" # change the start and end into "0" to make it a way
    #gmainmaze[dx][dy] = "0"
    
    findKey(sx,sy,ykx,yky)
    
    YkeyToYDoor(ykx,yky,ydx,ydy)
    
    DoorToKey(ydx,ydy,rkx,rky)

    RkeyToRDoor(rkx,rky,rdx,rdy)
    
    DoorToKey(rdx,rdy,bkx,bky)
    
    gmainmaze[rdx][rdy] = "0"
    gmainmaze[ydx][ydy] = "0"
    
    BkeyToBDoor(bkx,bky,bdx,bdy)
    
    DoorToKey(bdx,bdy,gkx,gky)
    
    GkeyToGDoor(gkx,gky,gdx,gdy)
    
    DoorToDestination(gdx,gdy,ex,ey)
    
    for i in backtracker:
        gmainmaze[i[0]][i[1]] = "s"

    for i in YdoorTracker:
        gmainmaze[i[0]][i[1]] = "s"
        
    for i in keyTracker:
        gmainmaze[i[0]][i[1]] = "s"
        
   
    
    for i in GdoorTracker:
        gmainmaze[i[0]][i[1]] = "s"
    
    for i in RdoorTracker:
        gmainmaze[i[0]][i[1]] = "s"
    
    for i in BdoorTracker:
        gmainmaze[i[0]][i[1]] = "s"
 
    for i in FinalTracker:
        gmainmaze[i[0]][i[1]] = "s"
    
    #print the maze
    width = len(gmainmaze[0]) - 1
    print("Check Your Map...")
    print("Map:")
    print("--"*(width+1))
    printMaze(gmainmaze)
    print("--"*(width+1))
    #visualize the maze uncomment to establish the function (btw check if your terminal can print these unicodes)
    visualize(gmainmaze)
    time.sleep(1)

main()
```
